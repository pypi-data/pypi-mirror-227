```python
from mrftools import *
import numpy as np
from matplotlib import pyplot as plt

SheppLogan_traditional = GenerateTraditionalPhantom(256, PhantomType.SHEPP_LOGAN)
ModifiedSheppLogan_traditional = GenerateTraditionalPhantom(256, PhantomType.MODIFIED_SHEPP_LOGAN)
plt.figure(figsize=(15, 3), dpi=100)
plt.subplot(1,2,1);plt.imshow(SheppLogan_traditional, cmap='gray', vmin=0, vmax=2); plt.colorbar(); plt.title("Shepp-Logan");
plt.subplot(1,2,2); plt.imshow(ModifiedSheppLogan_traditional, cmap='gray',vmin=0, vmax=1); plt.colorbar(); plt.title("Modified Shepp-Logan");

SheppLogan_3T = GenerateRelaxationPhantom(256, PhantomType.SHEPP_LOGAN, fieldStrengthTesla=3)
plt.figure(figsize=(15, 3), dpi=100)
plt.subplot(1,3,1);plt.imshow(SheppLogan_3T['T1'], cmap=T1_COLORMAP, vmin=T1_COLORMAP_MIN, vmax=T1_COLORMAP_MAX); plt.colorbar(); plt.title("T1 (3T)");
plt.subplot(1,3,2); plt.imshow(SheppLogan_3T['T2'], cmap=T2_COLORMAP, vmin=T2_COLORMAP_MIN, vmax=T2_COLORMAP_MAX); plt.colorbar(); plt.title("T2 (3T)");
plt.subplot(1,3,3);plt.imshow(SheppLogan_3T['PD'], cmap=B0_COLORMAP, vmin=B0_COLORMAP_MIN, vmax=B0_COLORMAP_MAX); plt.colorbar();  plt.title("PD (3T)");

SheppLogan_1p5T = GenerateRelaxationPhantom(256, PhantomType.SHEPP_LOGAN, fieldStrengthTesla=1.5)
plt.figure(figsize=(15, 3), dpi=100)
plt.subplot(1,3,1);plt.imshow(SheppLogan_1p5T['T1'], cmap=T1_COLORMAP, vmin=T1_COLORMAP_MIN, vmax=T1_COLORMAP_MAX); plt.colorbar(); plt.title("T1 (1.5T)");
plt.subplot(1,3,2); plt.imshow(SheppLogan_1p5T['T2'], cmap=T2_COLORMAP, vmin=T2_COLORMAP_MIN, vmax=T2_COLORMAP_MAX); plt.colorbar(); plt.title("T2 (1.5T)");
plt.subplot(1,3,3);plt.imshow(SheppLogan_1p5T['PD'], cmap=B0_COLORMAP, vmin=B0_COLORMAP_MIN, vmax=B0_COLORMAP_MAX); plt.colorbar();  plt.title("PD (1.5T)");
```
